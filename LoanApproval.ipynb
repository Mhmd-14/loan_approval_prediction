```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
df0 = pd.read_csv("loan_approval_dataset.csv")
df0
```

```python
df0.isna().sum()
```

```python
df0.rename(columns= {' loan_status':'loan_status',' self_employed':'self_employ'},inplace=True)
```

```python
x = df0.drop(['loan_status','loan_id',' education'],axis=1)
x
```

```python
y = df0[['loan_status']]
y
```

```python
z = pd.get_dummies(df0[' education'])
x = pd.concat([x,z],axis=1)
x
```

```python
x[['self_employ']]
```

```python
x['self_employ'] = x['self_employ'].str.replace('Yes','1')
x['self_employ'] = x['self_employ'].str.replace('No','0')
```

```python
x
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
model = LogisticRegression(random_state=50)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
mat = metrics.confusion_matrix(y_test,y_pred)
```

```python
mat
```

```python
print(f"The number of people who are actually approved for the loan and predicted correctly to be approved is:\t{mat[0][0]}")
```

```python
print(f"The number of people who are actually not approved for the loan and predicted correctly to be not approved is:\t{mat[1][1]}")
```
